```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head(10)
```

## Player Count

* Display the total number of players


```python
#Count number of unique screen names
count = purchase_data_df["SN"].nunique()
count
player_count_df = pd.DataFrame({"Total Players": [count]})

player_count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#find total items
items = purchase_data_df["Item Name"].nunique()
items

#find total spent
total_spent = purchase_data_df["Price"].sum()
total_spent

#find number of total purchases
total_purchases = purchase_data_df["Purchase ID"].nunique()
total_purchases

#calculate avg price per item
avg_price = total_spent / total_purchases
avg_price

#create summary data frame to display results
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [items],
                                      "Average Price": [avg_price],
                                      "Number of Purchases": [total_purchases],
                                      "Total Revenue": [total_spent]})
purchasing_analysis_df

```

```python
#summary data frame with formatting
final_pa_df = pd.DataFrame()

final_pa_df["Number of Unique Items"] = purchasing_analysis_df["Number of Unique Items"]
final_pa_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:.2f}".format)
final_pa_df["Number of Purchases"] = purchasing_analysis_df["Number of Purchases"]
final_pa_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:.2f}".format)
final_pa_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#remove duplicate records
drop_duplicates = purchase_data_df.drop_duplicates(["SN"], keep = "last")

#count number of m, f, and o / n-d
drop_duplicates["Gender"].value_counts()

#find number of males, and what percent of players
males = drop_duplicates.loc[drop_duplicates['Gender'] == "Male",:]

males_count = len(males["SN"].unique())

percent_male = round(males_count / count * 100, 2)

#find number of females and what percent of players
females = drop_duplicates.loc[drop_duplicates['Gender'] == "Female",:]

females_count = len(females["SN"].unique())

female_percent = round(females_count / count *100, 2)

#find number of other / non-disclosed and what percent of players
others = drop_duplicates.loc[drop_duplicates['Gender'] == "Other / Non-Disclosed",:]

others_count = len(others["SN"].unique())

others_percent = round(others_count / count * 100, 2)

#create dictionary to hold all of my libraries
gender_dict = {"Gender": ["Male", "Female", "Other / Non-Disclosed"],"Total Players": [males_count, females_count, others_count],
              "Percentage of Players": [percent_male, female_percent, others_percent]}

#build summary data frame to disply data
gender_demo_df = pd.DataFrame(gender_dict)
gender_demo_df = gender_demo_df.set_index("Gender")
gender_demo_df.index.names = [None]
gender_demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#count number of unique male ids
male_id = purchase_data_df.loc[purchase_data_df['Gender'] == "Male",:]

#number of purchases made by males
male_purchases = len(male_id["Purchase ID"].unique())

#total spent by male players
male_total_spent = round(male_id["Price"].sum(),2)

#avg spent by male players
male_avg_purchase = male_total_spent / male_purchases

#avg spent per male player
male_avg_per_person = male_total_spent / males_count

# number of female ids
female_id = purchase_data_df.loc[purchase_data_df['Gender'] == "Female",:]

# number of purchases by female players
female_purchases = len(female_id["Purchase ID"].unique())

#total spent by female players
female_total_spent = round(female_id["Price"].sum(), 2)

# avg spent by female players
female_avg_purchase = female_total_spent / female_purchases

# avg spent per female player
female_avg_per_person = female_total_spent / females_count

# number of other ids
others_id = purchase_data_df.loc[purchase_data_df['Gender'] == "Other / Non-Disclosed",:]

# number of purchases by other players
others_purchases = len(others_id["Purchase ID"].unique())

#total spent by other players
others_total_spent = round(others_id["Price"].sum(), 2)

# avg spent by other players
others_avg_purchase = others_total_spent / others_purchases

# avg spent per other player
others_avg_per_person = others_total_spent / others_count

# build summary data frame to display results
gender_purchase_analysis_df = pd.DataFrame({"Purchase Count": [male_purchases, female_purchases, others_purchases],
                                            "Gender": ["Male", "Female", "Other / Non-Disclosed"],
                                      "Average Purchase Price": [male_avg_purchase, female_avg_purchase, others_avg_purchase],
                                      "Total Purchase Value": [male_total_spent, female_total_spent, others_total_spent],
                                      "Avg Total Purchase per Person": [male_avg_per_person, female_avg_per_person, others_avg_per_person]})

gender_purchase_analysis_df = gender_purchase_analysis_df.set_index("Gender")

# format summary data frame for final results
gender_purchase_analysis_df["Average Purchase Price"] = gender_purchase_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
gender_purchase_analysis_df["Total Purchase Value"] = gender_purchase_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_analysis_df["Avg Total Purchase per Person"] = gender_purchase_analysis_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

# display final data frame
gender_purchase_analysis_df
```
